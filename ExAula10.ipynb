```python
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
vinho = datasets.load_wine()
```

```python
vinhos = pd.DataFrame(data = vinho.data, columns = vinho.feature_names)
vinhos['class'] = vinho.target
```

```python
vinhos
```

```python
x, y = vinhos.drop(columns=['class']), vinhos['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,
                                                    random_state = 0)
```

###Árvore Decisão:

```python
ad = tree.DecisionTreeClassifier(random_state = 42,
                                 max_depth = 4,
                                 criterion = 'entropy')
ad = ad.fit(x_train, y_train) #fitness, treino
```

```python
plt.figure(figsize=(10, 15))
tree.plot_tree(ad, fontsize = 10)
plt.show()
```
