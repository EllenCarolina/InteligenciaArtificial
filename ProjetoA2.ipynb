#Pré-Processamento

##Limpeza de base

###Verificação de NULL

```python
#Verifica todo df e responde True, se for nulo, ou False, se não-nulo
df_check = df.isnull()
```

```python
#Mostra quantos valores existem entre False e True, sendo valor absoluto
df_check.value_counts()
```

```python
#Mostra os valores em porcentagem
df_check.value_counts(normalize = True)
```

```python
#Retorna apenas os valores NULL da coluna, nam é pra erro, não necessariamente para elementos faltantes
df[df['salary'].isna()]
```

```python
#Dá dados estatísticos
df.describe()
```

###Seleção

```python
#deletar uma coluna desnecessária do DataFrame, salvando o anterior, residual, pra possível utilidade
df_res = df.drop(columns='Unnamed: 0')
```

```python
df_res.head()
```

```python
#Excluiu duas colunas desnecessária, no caso proposto
df_res = df_res.drop(columns=['salary_currency','salary'])
```

```python
df_res.head()
```

```python
#Conta os valores da tabela, o dropna conta quantos valores não foram retornados, ou seja não aplicável/erros
df_res['work_year'].value_counts(dropna = False)
```

```python
df_res['experience_level'].value_counts(dropna = False, normalize = True)
```
