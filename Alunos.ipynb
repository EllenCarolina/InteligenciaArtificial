```python
a3['Tem A?'] = 0
```

```python
for l in a3.T:
  a1 = a3.loc[l, 'Nome'].lower()
  if ((a1.find('a') > - 1) | (a1.find('ã') > - 1)):
    a3.loc[l, 'Tem A?'] = 'Tem'
  else:
    a3.loc[l, 'Tem A?'] = 'Não Tem'
```

```python
valor = a3.pop('Tem A?')
a3.insert(1, 'Tem A?', valor)
```

```python
a4 = a3.groupby('Idade')['Código Idade'].sum().reset_index().sort_values('Idade')
```

```python
#Gráfico de Barras Verticais
a4.plot.bar(
    x = 'Idade',
    y = 'Código Idade',
    rot = 0,
    figsize = (15, 5),
    title = 'Quantidade de Alunos por Idade'
)
```

```python
#Gráfico de Barras Horizontais:
a4.plot.barh(
    x = 'Idade',
    y = 'Código Idade',
    rot = 0,
    figsize = (10, 5),
    title = 'Quantidade de Alunos por Idade',
)
```
