```python
def questao(total_questoes: int)->list:
  gabarito_escopo_local = []
  for i in range(0, total_questoes, 1):
    print('Questão', (i + 1),': ')
    resposta = str(input())
    gabarito_escopo_local.append(resposta)
  return (gabarito_escopo_local)

gabarito = []
gabarito_aluno = []
nome = []
acertos = []
total_questoes = int(input('Quantas questões você deseja adicionar? '))

gabarito = questao(total_questoes)
alunos = int(input('Quantos alunos você deseja verificar a nota? '))
for l in range(0, alunos, 1):
  print('ALUNO ', (l + 1))
  print('-------')
  nome.append(str(input('Nome: ')))
  print('Respostas: ')
  gabarito_aluno = questao(total_questoes)
  correto = 0
  for index in range(0, len(gabarito), 1):
    if (gabarito_aluno[index] == gabarito[index]):
      correto = correto + 1
  acertos.append(correto)

for linha in range(0, alunos, 1):
  print('O(A) aluno(a)', nome[linha], 'acertou: ', acertos[linha], '/', total_questoes)
```
