```python
import pandas as pd
import numpy as np
import spacy
from spacy.lang.pt.examples import sentences 

import matplotlib.pyplot as plt
import seaborn as sns

import sklearn
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
ted = pd.read_csv('/content/data.csv')
```

```python
ted.head()
```

```python
array_de_df = []
for counter in range(len(ted)):
  ted_title = ted.loc[counter, 'title']
  array = []
  doc = nlp(ted_title)
  for token in doc:
    x = (token.text, token.lemma_, token.pos_)
    array.append(x)
  array_de_df.append(pd.DataFrame(array))
```

```python
array_de_df[16]
```

```python
for df in array_de_df:
  counter = 0
  while (counter < len(df)):
    if (df.loc[counter, 2] == 'PUNCT' or df.loc[counter, 2] == 'SYM' or df.loc[counter, 0] == ' '):
      df.drop(counter, inplace = True)
      df.reset_index(drop = True, inplace = True)
    else:
      counter = counter + 1
```

```python
array_de_df[16]
```

```python
array_de_palavras = []
[[array_de_palavras.append(palavra) for palavra in df[1]] for df in array_de_df]
print(array_de_palavras)
```

```python
for i in range(len(array_de_palavras)):
  if array_de_palavras[i] == 'need':
    print(i)
```

```python
unicas = np.unique(array_de_palavras)
```

```python
array_de_ids = array_de_palavras.copy()
for i_unica in range(len(unicas)):
  for i_array_de_palavras in range(len(array_de_palavras)):
    if array_de_palavras[i_array_de_palavras] == unicas[i_unica]:
      array_de_ids[i_array_de_palavras] = i_unica

print(array_de_ids)
```
