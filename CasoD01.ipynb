```python
sares.drop(8, inplace = True)
sares.reset_index(inplace = True)
sares.drop(columns={'level_0', 'index'}, inplace = True)
```

```python
for l in sares.T:
  ns = sares.loc[l, 'Nascimento']
  if (ns == 'YT, and perhaps firstborn'):
    sares.loc[l,'Nascimento'] = 'YT firstborn'
  elif (ns == 'Possibly pre First Age'):
    sares.loc[l, 'Nascimento'] = 'FA'
  if (ns == 'First Age'):
    sares.loc[l, 'Nascimento'] = 'FA'
  elif(ns == 'March 1 ,2931'):
    sares.loc[l, 'Nascimento'] = ns.replace('March 1 ,', 'TA ')
```

```python
for l in sares.T:
  ms = sares.loc[l, 'Morte']
  if (ms.find('departed') > - 1):
    sares.loc[l, 'Morte'] = 'Still Alive'
  elif (ms.find('Departed') > - 1):
    sares.loc[l, 'Morte'] = 'Still Alive'
  if (ms.find('still alive') > - 1):
    sares.loc[l, 'Morte'] = 'Still Alive'
```
