```python
import spacy
from spacy.lang.pt.examples import sentences
```

```python
import pandas as pd
```

```python
nlp = spacy.load('en_core_web_sm')
```

#Entrada da fonte de daos

```python
  doc = nlp('Gandalf: You should not pass, Microsoft!')
```

##Tokenização:

```python
for token in doc:
  print(token.text)
```

```python
texto = []
for token in doc:
  texto.append(token.text)
print(texto)
```

```python
pd.DataFrame(texto).T
```

```python
for token in doc:
  print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_, token.shape_, token.is_alpha, token.is_stop)
```

```python
texto_2 = []
for token in doc:
  x = (token.text, token.lemma_, token.pos_, token.tag_, token.dep_, token.shape_, token.is_alpha, token.is_stop)
  texto_2.append(x)
```

```python
pd.DataFrame(texto_2)
```

```python
tratamento = pd.DataFrame(texto_2)
```

```python
tratamento.rename(columns={0:'texto', 1:'lemma',
                           2:'propriedade_lexical',
                           3:'tag',
                           4:'relacao_depend.sintatica',
                           5:'qnt_caracters', 6:'boll_caracter_alfabe',
                           7:'stop_list'})
```

```python
for ent in doc.ents:
  print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
texto_3 = []
for ent in doc.ents:
  entidade = (ent.text, ent.start_char, ent.end_char, ent.label_)
  texto_3.append(entidade)
```

```python
entidade = pd.DataFrame(entidade).T
entidade.rename(columns={0:'texto', 1:'início', 2:'fim', 3:'rotulo'})
```
