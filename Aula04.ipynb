###Transposição/Pivoteamento:

```python
#linhas viram colunas
df_res_1.T
```

```python
df_res_1.set_index('rank').T
```

```python
df_melt = df[['directors', 'nome', 'rank', 'genre']]
```

```python
#Ordenei pelo rank
df_melt.sort_values(by = 'directors').sort_values('rank')
```

```python
pd.melt(df_melt, id_vars= ['directors', 'rank'], value_vars=['genre']).sort_values(by = 'rank')
```

```python
pd.melt(df_melt, id_vars= ['directors'], value_vars=['genre', 'rank']).sort_values(by = 'directors').head(2)
```
