```python
nlp = spacy.load('en_core_web_sm')
```

```python
ted = pd.read_csv('/content/data.csv')
```

```python
array_2d = []
for counter in ted.T:
  ted_title = ted.loc[counter, 'title']
  array = []
  doc = nlp(ted_title)
  for token in doc:
    array.append(token)
  array_2d.append(array) 
```
