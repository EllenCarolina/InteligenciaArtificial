##Função:

```python
def ajuste_tabela(df, Estado):
  df = df.rename(columns={'Unnamed: 0': 'Ano'})
  df = pd.melt(df, id_vars=['Ano'], value_vars=['Janeiro', 'Fevereiro',
                                                'Março', 'Abril', 'Maio',
                                                'Junho', 'Julho', 'Agosto',
                                                'Setembro', 'Outubro',
                                                'Novembro', 'Dezembro'],
               value_name = 'Número de Incêndios',
               var_name = 'Mês'
  )
  df['Estado'] = Estado
  return df    
```

```python
df_alagoas = ajuste_tabela(df_alagoas, 'Alagoas')
```

```python
df_alagoas
```

```python
df_amapa = ajuste_tabela(df_amapa,'amapa')
df_amazonas = ajuste_tabela(df_amazonas,'amazonas')
df_bahia = ajuste_tabela(df_bahia,'bahia')
df_ceara = ajuste_tabela(df_ceara,'ceara')
df_distrito_federal = ajuste_tabela(df_distrito_federal,'distrito_federal')
df_espirito_santo = ajuste_tabela(df_espirito_santo,'espirito_santo')
df_goias = ajuste_tabela(df_goias,'goias')
df_maranhao = ajuste_tabela(df_maranhao,'maranhao')
df_mato_grosso = ajuste_tabela(df_mato_grosso,'mato_grosso')
df_mato_grosso_do_sul = ajuste_tabela(df_mato_grosso_do_sul,'mato_grosso_do_sul')
df_para = ajuste_tabela(df_para,'para')
df_paraiba = ajuste_tabela(df_paraiba,'paraiba')
df_parana = ajuste_tabela(df_parana,'parana')
df_pernambuco = ajuste_tabela(df_pernambuco,'pernambuco')
df_minas_gerais = ajuste_tabela(df_minas_gerais,'minas_gerais')
df_piaui = ajuste_tabela(df_piaui,'piaui')
df_rio_de_janeiro = ajuste_tabela(df_rio_de_janeiro,'rio_de_janeiro')
df_rio_grande_do_norte = ajuste_tabela(df_rio_grande_do_norte,'rio_grande_do_norte')
df_rio_grande_do_sul = ajuste_tabela(df_rio_grande_do_sul,'rio_grande_do_sul')
df_rondonia = ajuste_tabela(df_rondonia,'rondonia')
df_roraima = ajuste_tabela(df_roraima,'roraima')
df_santa_catarina = ajuste_tabela(df_santa_catarina,'santa_catarina')
df_sergipe = ajuste_tabela(df_sergipe,'sergipe')
df_tocantins = ajuste_tabela(df_tocantins,'tocantins')
df_sp = ajuste_tabela(df_sp,'sp')
```
